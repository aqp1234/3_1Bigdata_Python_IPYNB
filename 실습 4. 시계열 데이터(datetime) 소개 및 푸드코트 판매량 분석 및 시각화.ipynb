## Tip. Week와 요일 정보 얻기(strftime)

```python
# 심화 - Week와 요일 정보 얻기
# 참고 사이트 : https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior

# %U : week number (sunday to monday) (e.g. 00, 01, ...53)
# %V : ISO 8601 week number (monday to sunday) (e.g. 01, 02, ... 53)
...
# %a : weekday (Sun, Mon, ..) 
# %w : weekday (0,1,2,3,...) (0 = Sunday)
```

```python
now.strftime('%V') # 주차(week number) 정보
```

```python
now.strftime('%a') # 요일 정보 (Sun, ... Mon)
```

```python
now.strftime('%w')# 요일 정보(0,1,2,...) 0 = sunday
```

```python
# isocalandar()도 사용 가능
# (년도, 주차, 요일) 정보 반환 
now.isocalendar() 
```

### (실습 #2) df에  week number(%V), weekday(%w) 컬럼 추가

```python
#df에  week number(%V), weekday(%w) 컬럼 추가
df['weekNumber'] = df['date2'].apply(lambda x: x.strftime('%V'))
df['weekDay'] = df['date2'].apply(lambda x: x.strftime('%w'))
df
```

```python
# 요일별 value의 합
df.pivot_table(index = 'weekDay', aggfunc = 'sum', values = 'value')
```

### date_range()를 활용한 고급 기법
https://pandas.pydata.org/pandas-docs/stable/generated/pandas.date_range.html

```python
# 특정 기간의 날짜 생성
pd.date_range(start='2018/01/25', end='2018/02/10')
```

```python
# start 날짜를 기준으로 periods 갯수만큼 날짜 생성
pd.date_range(start='2018/03/02', periods=7)
```

```python
#특정 주기로 날짜를 생성 - freq 인자
# 상세 내용은 https://pandas.pydata.org/pandas-docs/stable/timeseries.html#timeseries-offset-aliases 확인
pd.date_range(start='2018/03/02', end = '2018/09/01', freq='B') 

# D: 일단위
# W: 주 단위 (일요일만 선택)
# W-MON: 주 단위이되, 월요일만 선택
# W-TUE: 주 단위이되, 화요일만 선택. WED, THU, FRI, SAT, SUN 모두 가능
# M: 매달 마지막날만 선택
# MS: 매달 첫째날만 선택
# B: 주중만 선택 (Business day)
```
